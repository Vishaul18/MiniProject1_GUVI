```python
#library installation
import pandas as pd
import numpy as np
```

```python
#file reading
ds = pd.read_csv('/workspaces/MiniProject1_GUVI/Uber_Eats_data.csv')
ds.head()
```

```python
#droping unwanted columns
ds = ds.drop(columns = ['phone','rest_type','dish_liked','listed_in(type)','listed_in(city)'])
ds.head()
```

```python
ds.info()
```

```python
#data cleaning in name column
print(ds.name.isnull().sum())
ds.name = ds.name.str.lower().str.strip()
```

```python
#data cleaning in online_order column
print(ds.online_order.isnull().sum())
ds.online_order = ds.online_order.str.lower().str.strip()
```

```python
#data cleaning book_table
print(ds.book_table.isnull().sum())
ds.book_table = ds.book_table.str.lower().str.strip()
```

```python
#data cleaning rate
ds.rate = ds.rate.str.strip()
ds['rate'] = ds['rate'].str.replace('/5', '', regex=False)
ds['rate'] = pd.to_numeric(ds['rate'], errors='coerce')
```

```python
ds.info()
```

```python
#checking null values in rate column
ds.rate.isnull().sum()
```

```python
#dropping null values in rate column
ds = ds.dropna(subset=['rate'])
```

```python
ds.info()
```

```python
# lowercase and extra space removal in location column data
ds.location = ds.location.str.lower().str.strip()
```

```python
# null value check in location column
ds.location.isnull().sum()
```

```python
# lowercase and extra space removal in cuisines column data
ds.cuisines = ds.cuisines.str.lower().str.strip()
```

```python
# null value check in cuisines column
ds.cuisines.isnull().sum()
```

```python
#creation of cost_for_two columns and cilling data from approx_cost(for two people) column
ds['cost_for_two'] = ds['approx_cost(for two people)']
```

```python
ds.info()
```

```python
#droping approx_cost(for two people) column
ds = ds.drop(columns = 'approx_cost(for two people)')
```

```python
#removing the extra spaces and , replacement
ds['cost_for_two'] = ds['cost_for_two'].str.strip()
ds['cost_for_two'] = ds['cost_for_two'].str.replace(',', '')
ds['cost_for_two'] = pd.to_numeric(ds['cost_for_two'], errors='coerce')
```

```python
# checking for null values
ds.cost_for_two.isnull().sum()
```

```python
ds.info()
```

```python
#rating normalization
C = ds['rate'].mean()
m = ds['votes'].quantile(0.75)

ds['normalized_rating'] = ((ds['votes'] / (ds['votes'] + m)) * ds['rate'] +(m / (ds['votes'] + m)) * C)
```

```python
ds.info()
```

```python
ds.duplicated().sum()
```

```python
ds.duplicated(subset=['name', 'location']).sum()
```

```python
duplicates = ds[ds.duplicated(subset=['name', 'location'], keep=False)]
duplicates
```

```python
# droping duplicates based on name and location
ds = ds.drop_duplicates(subset = ['name','location'])
```

```python
ds.info()
```

```python
ds.to_csv('/workspaces/MiniProject1_GUVI/cleaned_uber_eats.csv', index=False)
```

```python
query = """
SELECT 
    location,

    CASE 
        WHEN cost_for_two < 500 THEN 'Low'
        WHEN cost_for_two BETWEEN 500 AND 1000 THEN 'Mid'
        ELSE 'High'
    END AS price_category,

    COUNT(*) AS total_restaurants,
    ROUND(AVG(normalized_rating), 2) AS avg_rating

FROM restaurants

GROUP BY location, price_category

HAVING COUNT(*) >= 5

ORDER BY avg_rating DESC
LIMIT 10;
"""
mycursor.execute(query)
out = mycursor.fetchall()
print(tabulate(out, headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```

```python
query = """
SELECT 
    name,
    location,
    normalized_rating,

    CASE 
        WHEN cost_for_two < 500 THEN 'Low'
        WHEN cost_for_two BETWEEN 500 AND 1000 THEN 'Mid'
        ELSE 'High'
    END AS price_category

FROM restaurants

ORDER BY price_category, normalized_rating DESC;
"""
mycursor.execute(query)
out = mycursor.fetchall()
print(tabulate(out[:20], headers=[i[0] for i in mycursor.description], tablefmt='psql'))
```
